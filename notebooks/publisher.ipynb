```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file
data = pd.read_csv('C:/Users/hp/Desktop/week1/data/raw_analyst_ratings.csv')
```

```python

# Define categories and associated keywords
categories = {
    'Sports': ['football', 'soccer', 'championship', 'game', 'match'],
    'Politics': ['government', 'policy', 'election', 'politics'],
    'Entertainment': ['celebrity', 'movie', 'film', 'show', 'music'],
    'Business': ['stock', 'market', 'economy', 'business', 'finance']
}

def categorize_headline(headline):
    headline_lower = headline.lower()
    for category, keywords in categories.items():
        if any(keyword in headline_lower for keyword in keywords):
            return category
    return 'Other'  # Return 'Other' if no keywords match

# Apply the categorization function to the DataFrame
data['category'] = data['headline'].apply(categorize_headline)

# Count the total number of contributions by each publisher
publisher_counts = data['publisher'].value_counts()

# Print the counts for each publisher
print("Total contributions by each publisher:")
print(publisher_counts)

# Analyze the number of publications by each publisher and category (type of news)
publisher_category_counts = data.groupby(['publisher', 'category']).size().unstack()

# Print the count of publications by publisher and category
print("\nNumber of publications by publisher and category:")
print(publisher_category_counts)

# Get the top 20 publishers by number of contributions
top_publishers = publisher_counts.head(20).index

# Filter the dataset to include only the top 20 publishers
df_top_publishers = data[data['publisher'].isin(top_publishers)]

# Print the number of contributions by the top 20 publishers
print("\nContributions by top 20 publishers:")
print(df_top_publishers['publisher'].value_counts())

# Analyze the number of publications by the top 20 publishers and category (type of news)
top_publisher_category_counts = df_top_publishers.groupby(['publisher', 'category']).size().unstack()

# Print the count of publications by top 20 publishers and category
print("\nNumber of publications by top 20 publishers and category:")
print(top_publisher_category_counts)
```

```python

# Define categories and associated keywords
categories = {
    'Sports': ['football', 'soccer', 'championship', 'game', 'match'],
    'Politics': ['government', 'policy', 'election', 'politics'],
    'Entertainment': ['celebrity', 'movie', 'film', 'show', 'music'],
    'Business': ['stock', 'market', 'economy', 'business', 'finance']
}

def categorize_headline(headline):
    headline_lower = headline.lower()
    for category, keywords in categories.items():
        if any(keyword in headline_lower for keyword in keywords):
            return category
    return 'Other'  # Return 'Other' if no keywords match

def extract_domain(email):
    return email.split('@')[1] if '@' in email else 'Unknown'

# Apply the categorization function to the DataFrame
data['category'] = data['headline'].apply(categorize_headline)

# Extract domains from publisher emails
data['domain'] = data['publisher'].apply(extract_domain)

# Count the total number of contributions by each publisher
publisher_counts = data['publisher'].value_counts()

# Print the counts for each publisher
print("Total contributions by each publisher:")
print(publisher_counts)

# Count the total number of contributions by each domain
domain_counts = data['domain'].value_counts()

# Print the counts for each domain
print("\nTotal contributions by each domain:")
print(domain_counts)

# Analyze the number of publications by each publisher and category (type of news)
publisher_category_counts = data.groupby(['publisher', 'category']).size().unstack()

# Print the count of publications by publisher and category
print("\nNumber of publications by publisher and category:")
print(publisher_category_counts)

# Get the top 20 publishers by number of contributions
top_publishers = publisher_counts.head(20).index

# Filter the dataset to include only the top 20 publishers
df_top_publishers = data[data['publisher'].isin(top_publishers)]

# Print the number of contributions by the top 20 publishers
print("\nContributions by top 20 publishers:")
print(df_top_publishers['publisher'].value_counts())

# Analyze the number of publications by the top 20 publishers and category (type of news)
top_publisher_category_counts = df_top_publishers.groupby(['publisher', 'category']).size().unstack()

# Print the count of publications by top 20 publishers and category
print("\nNumber of publications by top 20 publishers and category:")
print(top_publisher_category_counts)

# Analyze the number of publications by each domain and category (type of news)
domain_category_counts = data.groupby(['domain', 'category']).size().unstack()

# Print the count of publications by domain and category
print("\nNumber of publications by domain and category:")
print(domain_category_counts)
```
